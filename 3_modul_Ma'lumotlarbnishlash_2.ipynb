```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/world_population.csv", index_col='rank')
df.head()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
  rank INT, 
  country TEXT PRIMARY KEY,
  density Double, 
  population Double not null,
  area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
uz = df[df.country=="Uzbekistan"]
uz
```

```python
uz_rank = df[df.index==133]
uz_rank
```

```python
rank = uz.index.values[0]
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
population = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
print(f'{rank}, {country}, {density}, {population}, {area}')
```

```python
query = f"""
INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area} )
"""

cursor.execute(query)
db.commit()
```

```python
query = 'select country, year, coal, gas, hydro, oil  from elect  where country="Uzbekistan" and year>2010'
dfuz= pd.read_sql(query, db)
dfuz
```

```python
query = 'select country, year, coal, gas, hydro, oil  from elect  where country="Uzbekistan" and year between 2010 and 2015' # between
pd.read_sql(query, db)
```

```python
query = 'select country, year, coal, gas, hydro, oil  from elect  where country="Uzbekistan" and year in (2010, 2013, 2015)' # between
pd.read_sql(query, db)
```

```python
# LIKE
# %var - Anvar, Sarvar
# _ar  - car, bar, tar
# __ar - star
# %ol% - bola, xola, bolalar

query = "select * from elect where country like '%stan' and year in (2019, 2020, 2021)"
pd.read_sql(query, db)
```

```python
query = "select * from elect where code like 'A__' and year in (2020, 2021)"
pd.read_sql(query, db)
```
