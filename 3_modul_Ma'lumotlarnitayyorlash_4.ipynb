```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
range = pd.cut(df.price, bins)
range
```

```python
range.value_counts()
```

```python
names = ['cheap', 'Middle', 'Expensive']
range2 = pd.cut(df.price, bins, labels=names)
range2.value_counts()
```

```python
a = pd.cut(df.price, 3) # automatic bo'ladi.
a.value_counts()
```

```python
b = pd.qcut(df.price, 3) # catigoriyalar sonini nisbatan teng automatic bo'ladi. 
b.value_counts()
```
