```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df
```

```python
df.info()
```

```python
columns = ["City", "Region", "Country", "AirQuality", "WaterPollution"]
df.columns =columns 
```

```python
df.head()
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"', ""))
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"', ""))
df.head()
```

```python
df.to_csv("Updated_dataset.csv", index_label=False)
```

pivot_table orqali irarxik indekslash

```python
df.index
```

```python
df1 = pd.pivot_table(df, index = ['Country', "City"])
df1
```

```python
df1.loc[['Uzbekistan', 'Germany']]
```

```python
df2 = df1.copy()
df2
```

```python
df2["AirPollution"] = 100 - df2['AirQuality'] 
df2
```

```python
df2["WaterPollution"] = 100 - df2['WaterPollution'] 
df2["WaterQuality"] = 100 - df2['WaterPollution'] 
df2
```

```python
df3 = pd.DataFrame(df2.values, index = df2.index, columns = [["AIR", "AIR", 'WATER', 'WATER'], ['AirQuality', 'AirPollution', "WaterQuality", "WaterPollution"]])
df3
```

```python
df3.columns.names = ["Object", "Info"]
df3
```

```python
df3.index.names
```

```python
df3.swaplevel('Country', 'City') #irarxik indekslarni o'rnini almashtirishadi.
```
